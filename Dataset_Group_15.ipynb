```python
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf
from statsmodels.stats.outliers_influence import variance_inflation_factor
import numpy as np
import pandas as pd
```

```python
!git clone https://github.com/nikkhot/DTSC-Group-15
%cd DTSC-Group-15
df = pd.read_csv("College_Data_dataset.csv")
X1=df["Expend"]
X2=df["Outstate"]
X3=df["Books"]
X4=df["Personal"]
X5= df["Room.Board"]
Y = df['Enroll']
```

```python
df2= pd.DataFrame( {"X1":X1, "X2":X2, "X3":X3, "X4":X4, "X5":X5, "Y":Y })
```

```python
df2.head()
```

```python
plt.figure(figsize= (5,3))
sns.heatmap(df2.corr(), annot= True , cmap ="coolwarm", vmin=-1, vmax= 1)
plt.title("Correlation Matrix")
plt.show()
```
