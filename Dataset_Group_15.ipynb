```python
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf
from statsmodels.stats.outliers_influence import variance_inflation_factor
import numpy as np
import pandas as pd
```

```python
!git clone https://github.com/nikkhot/DTSC-Group-15
%cd DTSC-Group-15
df = pd.read_csv("College_Data_dataset.csv")
X1=df["Expend"]
X2=df["Outstate"]
X3=df["Books"]
X4=df["Personal"]
X5= df["Room.Board"]
Y = df['Enroll']
```

```python
df2= pd.DataFrame( {"Expend":X1, "Outstate":X2, "Books":X3, "Personal":X4, "Room.Board":X5, "Enroll":Y })
```

```python
df2.head()
```

```python
plt.figure(figsize= (5,3))
sns.heatmap(df2.corr(), annot= True , cmap ="coolwarm", vmin=-1, vmax= 1)
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.pairplot(df2, height=1.5)
plt.show()
```

```python
vif_data = pd.DataFrame()
X = df2[ ["Expend", "Outstate", "Books","Personal","Room.Board"]]
vif_data["Feature"] = X.columns

vif_data["VIF"] = [ variance_inflation_factor(X.values ,i) for i in range(X.shape[1] )  ]
vif_data
```
